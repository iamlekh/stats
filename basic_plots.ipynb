### MATPLOTLIB BASICS

#### Line Plot

```python
from numpy import sin
from matplotlib import pyplot
%matplotlib inline

x = [x*0.1 for x in range(100)]

y = sin(x)

pyplot.plot(x, y)
pyplot.show()
```

#### Bar Chart

```python
from random import seed
from random import randint
seed(4)

x = [ ' red ' , ' green ' , ' blue ' ]

y = [randint(0, 100), randint(0, 100), randint(0, 100)]

pyplot.bar(x, y)
pyplot.show()
```

#### Histogram Plot

```python
x = randn(1000)

pyplot.hist(x)
pyplot.show()
```

#### Box and Whisker Plot

```python
x = [5* randn(1000), 10 * randn(1000), 20 * randn(1000)]

pyplot.boxplot(x)
pyplot.show()
```

#### Scatter Plot

```python
x = 20 * randn(1000) + 100

y = x + (10 * randn(1000) + 50)

pyplot.scatter(x, y)
pyplot.show()
```
