# Continuous Distribution

#### Gaussian Distribution

```python
# plot the gaussian pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
%matplotlib inline
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
mean = 0.0
stdev = 1.0
# calculate the pdf
pdf = norm.pdf(sample_space, mean, stdev)
# plot
pyplot.plot(sample_space, pdf,lw=5)
pyplot.show()
```

```python
# plot the gaussian cdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
# calculate the cdf
cdf = norm.cdf(sample_space)
# plot
pyplot.plot(sample_space, cdf,lw=5)
pyplot.show()
```

#### Student’s t-Distribution

```python
# plot the t-distribution pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1
# calculate the pdf
pdf = t.pdf(sample_space, dof)
# plot
pyplot.plot(sample_space, pdf,lw=5)
pyplot.show()
```

```python
# plot the t-distribution cdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1
# calculate the cdf
cdf = t.cdf(sample_space, dof)
# plot
pyplot.plot(sample_space, cdf,lw=5)
pyplot.show()
```

#### Chi-Squared Distribution

```python
# plot the chi-squared pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2
# define the distribution parameters
sample_space = arange(0, 50, 0.01)
dof = 20
# calculate the pdf
pdf = chi2.pdf(sample_space, dof)
# plot
pyplot.plot(sample_space, pdf,lw=5)
pyplot.show()
```

```python
# plot the chi-squared cdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2
# define the distribution parameters
sample_space = arange(0, 50, 0.01)
dof = 20
# calculate the cdf
cdf = chi2.cdf(sample_space, dof)
# plot
pyplot.plot(sample_space, cdf,lw=5)
pyplot.show()
```

#### Exponential Distribution

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(expon.ppf(0.01),expon.ppf(0.99), 100)
pyplot.plot(x, expon.pdf(x) ,lw=5)
pyplot.show()
```

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(expon.ppf(0.01),expon.ppf(0.99), 100)
pyplot.plot(x, expon.cdf(x) ,lw=5)
pyplot.show()
```

#### Logistic Distriburtion

```python
from scipy.stats import logistic
import matplotlib.pyplot as plt
x = np.linspace(logistic.ppf(0.01),logistic.ppf(0.99), 100)
plt.plot(x, logistic.pdf(x),lw=5)
```

```python
from scipy.stats import logistic
import matplotlib.pyplot as plt
x = np.linspace(logistic.ppf(0.01),logistic.ppf(0.99), 100)
plt.plot(x, logistic.cdf(x),lw=5)
```

# Discrete Distribution

#### uniform distribution

```python
from scipy.stats import uniform
n = 100
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
plt.plot(data_uniform,lw=5)
plt.show()
```

```python
plt.hist(data_uniform)
plt.show()
```

#### Bernoulli Distribution

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)
plt.hist(data_bern)
plt.show()
```

#### Binomial Distribution

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
plt.hist(data_binom ,rwidth=0.9 , bins =10)
plt.show()
```

#### Poisson Distribution

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=5, size=1000)
plt.hist(data_poisson ,rwidth=0.9 , bins =10)
plt.show()
```
